```python
import numpy as np

python_list = [10, 20, 30, 40, 50]
array = np.array(python_list)
print("NumPy Array:", array)
```

```python
zeros_array = np.zeros((3, 4))  # 3x4 array of zeros
ones_array = np.ones((2, 5))    # 2x5 array of ones

print("Zeros Array:\n", zeros_array)
print("Ones Array:\n", ones_array)
```

```python
array = np.array([1, 2, 3, 4])
added = array + 5
multiplied = array * 3

print("Original Array:", array)
print("After Addition:", added)
print("After Multiplication:", multiplied)
```

```python
import numpy as np
array = np.array([10, 20, 30, 40, 50, 60])

print(array[:1])
print([2])
```

```python
import numpy as np
array_2d = np.array([[1, 2], [3, 4], [5, 6]])
transposed = array_2d.T

print("Original 2D Array:\n", array_2d)

print("Transposed Array:\n", transposed)
```

```python
import numpy as np
array = np.array([1, 4, 9, 16])
sqrt_array = np.sqrt(array)
exp_array = np.exp(array)
print("Original array:",array)
print("Square Roots:", sqrt_array)
print("Exponentials:", exp_array)
```

```python
import numpy as np
array = np.array([1, 2, 2, 3, 4, 4, 4, 5])

print("Maximum:", np.max(array))
print("Minimum:", np.min(array))
print("Unique Elements:", np.unique(array))
print("Bincount:", np.bincount(array))
```

```python
import numpy as np
array = np.array([[1, 2, 3], [4, 5, 6]])

print("Shape:", array.shape)
print("Data Type:", array.dtype)
print("Dimensions:", array.ndim)
print("Total Elements:", array.size)
```
