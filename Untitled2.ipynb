```python
import pandas as pd
s = pd.Series([1, 3, 5, 7, 9], name = 'numbers')
print("Series:\n",s)
data = {'Name' : ['alice', 'bob', 'Charlie'], 'Age' :[25, 30, 35], 'city' : ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data)
print("\nDataFrame:\n",df)

print("\nAge column:\n",df['Age'])
print("\nPeople older than 28\n",df[df['Age'] > 28])
```
