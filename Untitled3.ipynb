```python
import pandas as pd
import numpy as np
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily',
             'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no',
                'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=labels)
print("DataFrame:")
print(df)
```

```python
df.loc[df['name'] == 'James', 'name'] = 'Suresh'
print("After changing James to Suresh:")
print(df)
```

```python
df['age'] = [20, 21, 19, 22, 20, 23, 24, 21, 22, 20]
print("DataFrame after adding 'age' column:")
print(df)
```

```python
columns_list = df.columns.tolist()
print("List of column headers:", columns_list)
```

```python
import pandas as pd

s = pd.Series([1, 3, 5, 7, 9], name='numbers')
print("Series:\n", s)

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'London', 'Paris']}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)

print("\nAge column:\n", df['Age'])

print("\nPeople older than 28:\n", df[df['Age'] > 28])
```
